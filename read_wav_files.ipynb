```python
# Import modules and packages
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import librosa as lr
```

```python
# Set directory for source files
data_dir = './SIREN_SOUNDS/new_sirens'
audio_files = glob(data_dir + '/*.wav')
```

```python
# Perform reading through all the audio files
for file in range(0, len(audio_files), 1):
    
    # Read the first audio file, create the time array (timeline)
    audio, sfreq = lr.load(audio_files[file])
    time = np.arange(0, len(audio)) / sfreq
    
    # Plot audio over time
    fig, ax = plt.subplots()
    ax.plot(time, audio)
    ax.set(xlabel = 'Time', ylabel = 'Sound Amplitude')
    plt.show()
```
